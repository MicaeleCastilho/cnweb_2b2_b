# Formula da Velocidade Média

```
def calcVelMedia(deltaS, deltaT):
  velMedia = (deltaS / deltaT)
  return velMedia
```

teste de serviço: velocidade media

```
print('Velocidade Média: ',calcVelMedia(50, 4))
```

# Fórmula da aceleração

```
def calcularVelMedia (velocidade, tempo):
  VelMedia = (velocidade / tempo)
  return VelMedia
```

```
print("Aceleração =", calcularVelMedia(15, 2))
```

## Equação de Torricelli para lançamento vertical

```
def calcularTorricelliVer(vzero, gravi, alt):
  vFinal1 = (vzero ** 2) - (2 * gravi* alt)
  vFinal2 = (vzero ** 2) + (2 * gravi* alt)
  return vFinal1, vFinal2
```

```
print("Torricelli vertica [subindo], [descendo]", calcularTorricelliVer(50, 10, 100))
```
