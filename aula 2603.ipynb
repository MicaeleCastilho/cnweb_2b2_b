# Sistema de equações

```
import math

def eq2grau(a, b, c):
  delta = b**2 - 4*a*c

  if(delta == 0):
    return 'equação sem solução'
  else:
      x1 =(-b + math.sqrt(delta)) / (2*a)
      x2 =(-b - math.sqrt(delta)) / (2*a)
      return x1, x2

```

```
print(eq2grau(1, -3, 2))
```
